```python
def round_score():
    shape_score = {
        'A': 1, 
        'B': 2, 
        'C': 3, 
        'X': 1, 
        'Y': 2, 
        'Z': 3, 
    }
    
    # A, X for Rock, 
    # B, Y for Paper, 
    # C, Z for Scissors
    def get_points(o, s):
        if o == 'A' and s == 'X': 
            sp, op = 3, 3
        elif o == 'A' and s == 'Y':
            op = 0
            sp = 6
        elif o == 'A' and s == 'Z':
            op = 6
            sp = 0
        elif o == 'B' and s == 'X':
            op = 6
            sp = 0
        elif o == 'B' and s == 'Y': 
            sp, op = 3, 3
        elif o == 'B' and s == 'Z':
            op = 0
            sp = 6
        elif o == 'C' and s == 'X':
            op = 0
            sp = 6
        elif o == 'C' and s == 'Y':
            op = 6
            sp = 0
        elif o == 'C' and s == 'Z':
            sp, op = 3, 3
        return sp + shape_score[s], op + shape_score[o]
        

    with open(r'./input.txt', 'r') as f:
        shapes = f.readlines()
    
    other = [shape.split(' ')[0] for shape in shapes]
    self = [shape.strip('\n').split(' ')[1] for shape in shapes]
    rounds = len(other)
    
    total_self = 0
    total_other = 0
    for round in range(rounds):
        self_point, other_point = get_points(other[round], self[round])
        total_self += self_point
        total_other += other_point

    return total_other, total_self
```
